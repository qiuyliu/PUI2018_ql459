# Use API and download a json file

```python
url = "https://data.cityofnewyork.us/resource/pqfs-mqru.json"
```

```python
#use the appropriatepandas function to read in the json file
df = pd.read_json(url)
```

```python
#printout
df.head(3)
```

```python
df.tail(3)
```

```python
#now select a date and 2 variables
gt_3c = df[['lpep_pickup_datetime','fare_amount','trip_distance']]
```

```python
gt_3c.rename(columns = {'lpep_pickup_datetime':'Pickup Time',
                       'fare_amount':'Fare Amount',
                       'trip_distance':'Trip Distance'}, inplace = True)
```

```python
ax = gt_3c.plot(x='Pickup Time', y='Fare Amount');
ax.set_ylabel("Fare Amount");
```

The formatting of the datetime is hard to tell the timeline. Use the timestamp to formate the datetime to present.

```python
gt_3c['Pickup Time'] = gt_3c['Pickup Time'].apply(lambda x:pd.Timestamp(x))
```

```python
ax = gt_3c.plot(x='Pickup Time', y='Fare Amount')
ax.set_ylabel("Amount");
```

Figure 2: The fare amount of trip of Green taxi in NYC as a function of time on date 12/5/2016. The far amount and the pickup time is a negative correlation. The fare amount of the first minute turns lower through the timeline, but the range of fare amount is large.

```python
ax = gt_3c.plot(x='Pickup Time', y='Trip Distance')
ax.set_ylabel("Distance");
```

Figure 3: The trip distance of Green taxi in NYC as a function of time on date 12/5/2016. The trip distace and the pick up time is a negative correlation in first minute. The range of trip distance turns smaller.
