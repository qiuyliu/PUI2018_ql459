```python
def getCitiBikeCSV(datestring):
    print ("Downloading", datestring)
    ### First I will heck that it is not already there
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.csv"):
        if os.path.isfile(datestring + "-citibike-tripdata.csv"):
            # if in the current dir just move it
            if os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA")):
                print ("Error moving file!, Please check!")
        #otherwise start looking for the zip file
        else:
            if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip"):
                if not os.path.isfile(datestring + "-citibike-tripdata.zip"):
                    os.system("curl -O https://s3.amazonaws.com/tripdata/" + datestring + "-citibike-tripdata.zip")
                ###  To move it I use the os.system() functions to run bash commands with arguments
                os.system("mv " + datestring + "-citibike-tripdata.zip " + os.getenv("PUIDATA"))
            ### unzip the csv 
            os.system("unzip " + os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip")
            ## NOTE: old csv citibike data had a different name structure. 
            if '2014' in datestring:
                os.system("mv " + datestring[:4] + '-' +  datestring[4:] + 
                          "\ -\ Citi\ Bike\ trip\ data.csv " + datestring + "-citibike-tripdata.csv")
            os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA"))
    ### One final check:
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.csv"):
        print ("WARNING!!! something is wrong: the file is not there!")

    else:
        print ("file in place, you can continue")
```

# Trip duration of bikers that ride during the day vs night
### KS Tests
#### H0 : The daytime and night have the common distribution on trip duration
#### H1:  The daytime and night have the different distribution on trip duration
$$ \alpha = 0.05 $$

```python
df_dur['hour'] = df_dur['date'].dt.hour
```

```python
df_dur.head()
```

```python
#daytime 6am - 6pm
df_dur['daytime_trip'] = np.where((df_dur['hour']>=6) & (df_dur['hour'] < 18), 'daytime','night')
df_dur.head()
```

```python
df_dur_day = df[df_dur.daytime_trip == 'daytime']
df_dur_night = df[df_dur.daytime_trip == 'night']
```

```python
daytime_val = df_dur_day['tripduration'].values
night_val = df_dur_night['tripduration'].values
num_day = len(daytime_val)
num_night = len(night_val)
```

## KS tests to compare 2 samples

```python
ks = scipy.stats.ks_2samp(daytime_val, night_val)
print(ks)
```

### result: Null hypothesis is rejected, pvalue is smaller than the significance levle of 0.05; We accept the alternative hypothesis

## N/200

```python
subset_day = int ((len(df_dur_day)) / 200)
subset_night = int ((len(df_dur_night)) / 200)

df_subset_day = df_dur_day.sample(subset_day)
df_subset_night = df_dur_night.sample(subset_night)
#print(len(df_subset_day) + len(df_subset_night))
dt_val = df_subset_day['tripduration'].values
nt_val = df_subset_night['tripduration'].values
```

```python
scipy.stats.ks_2samp(dt_val, nt_val)
```

### result: using sample size in N/200, the p value is getting larger, 0.013 but still smaller than 0.05. The null hypothesis is still rejected.

## Testing for Correlation
### Null hypothesis: The trip duration's distribution of rides have linear corrleation in daytime and night
## Pearson's test for correlation

```python
np.random.seed(99)
cor_day = np.random.choice(df_dur_day['tripduration'], size=10000)
cor_night = np.random.choice(df_dur_night['tripduration'], size=10000)
sorted_day = np.sort(cor_day)
sorted_night = np.sort(cor_night)
```

```python
scipy.stats.pearsonr(sorted_day, sorted_night)
```

### result: The p value is 0, which is smaller than 5% level, so we could reject the null hypothesis. We can conclude that there are differences between distributions of trip duration by daytime and night

## Spearman's Test

```python
scipy.stats.spearmanr(sorted_day, sorted_night)
```

### result: The p value is 0, which is smaller than 5% level, so we could reject the null hypothesis. Moreover, the correlation result from spearman's test, 0.99, which is very close to 1, indicating a good correlation
